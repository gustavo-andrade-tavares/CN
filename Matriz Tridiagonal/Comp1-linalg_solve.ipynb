## Comparação utilizando linalg.solve

```python
import numpy as np
import timeit

def linalg_solve(n):
    d = 5
    a = -2

    e = np.ones(n)

    A = np.diag(d*e) + np.diag(np.array(a*e[range(n-1)]),1) + np.diag(np.array(a*e[range(n-1)]),-1)

    b = (d + (2*a))*e

    S = np.linalg.solve(A,b)
    return(S)

t = timeit.Timer("linalg_solve(64)", "from __main__ import linalg_solve")
print(t.repeat(5,1))
u = timeit.Timer("linalg_solve(128)", "from __main__ import linalg_solve")
print(u.repeat(5,1))
v = timeit.Timer("linalg_solve(256)", "from __main__ import linalg_solve")
print(v.repeat(5,1))
w = timeit.Timer("linalg_solve(512)", "from __main__ import linalg_solve")
print(w.repeat(5,1))
x = timeit.Timer("linalg_solve(1024)", "from __main__ import linalg_solve")
print(x.repeat(5,1))
y = timeit.Timer("linalg_solve(2048)", "from __main__ import linalg_solve")
print(y.repeat(5,1))
```
